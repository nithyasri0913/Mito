```python
from mitosheet import *; register_analysis("id-gclzcyxxac");
    
# Imported players_22.csv
import pandas as pd
players_22 = pd.read_csv(r'players_22.csv')

# Sorted overall in descending order
players_22 = players_22.sort_values(by='overall', ascending=False, na_position='last')

# Added column new-column-hmu6
players_22.insert(6, 'new-column-hmu6', 0)

# Reordered column new-column-hmu6
players_22_columns = [col for col in players_22.columns if col != 'new-column-hmu6']
players_22_columns.insert(13, 'new-column-hmu6')
players_22 = players_22[players_22_columns]

# Reordered column new-column-hmu6
players_22_columns = [col for col in players_22.columns if col != 'new-column-hmu6']
players_22_columns.insert(12, 'new-column-hmu6')
players_22 = players_22[players_22_columns]

# Renamed columns height_m
players_22.rename(columns={'new-column-hmu6': 'height_m'}, inplace=True)

# Deleted columns height_m
players_22.drop(['height_m'], axis=1, inplace=True)

# Added column new-column-2lpo
players_22.insert(12, 'new-column-2lpo', 0)

# Renamed columns height_m
players_22.rename(columns={'new-column-2lpo': 'height_m'}, inplace=True)

# Set formula of height_m
players_22['height_m'] = players_22['height_cm']/100

# Added column new-column-8ri8
players_22.insert(14, 'new-column-8ri8', 0)

# Renamed columns bmi
players_22.rename(columns={'new-column-8ri8': 'bmi'}, inplace=True)

# Set formula of bmi
players_22['bmi'] = players_22['weight_kg']/(players_22['height_m']*players_22['height_m'])

# Added column new-column-80ky
players_22.insert(15, 'new-column-80ky', 0)

# Deleted columns new-column-80ky
players_22.drop(['new-column-80ky'], axis=1, inplace=True)

# Changed league_name to dtype str
players_22['league_name'] = players_22['league_name']
```

# Create a Pivot Table


```python
#which countries have the tallest players?
# Pivoted into players_22
tmp_df = players_22[['nationality_name', 'height_cm']]
pivot_table = tmp_df.pivot_table(
    index=['nationality_name'],
    values=['height_cm'],
    aggfunc={'height_cm': ['mean']}
)
pivot_table.set_axis([flatten_column_header(col) for col in pivot_table.keys()], axis=1, inplace=True)
players_22_pivot = pivot_table.reset_index()

# Sorted height_cm mean in descending order
players_22_pivot = players_22_pivot.sort_values(by='height_cm mean', ascending=False, na_position='last')

# Filtered height_cm mean
players_22_pivot = players_22_pivot[players_22_pivot['height_cm mean'] > 180]

players_22_pivot
```
